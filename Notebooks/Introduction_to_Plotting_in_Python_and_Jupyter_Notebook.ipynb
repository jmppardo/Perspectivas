## Defining a function to plot

As our first example we are plotting the graph of the function defined by $f(t) = (t^2 + \frac{1}{2})\exp(-t^2)$ for values of $t$ between $-3$ and $3$. The first thing that we will need is a python function that is able to evaluate such an expression, that is, to calculate the values of the function for the corresponding values of $t$.

```python
def f(t):

    y = (t**2+0.5)*np.exp(-t**2)
    
    return y
```

This is a simple Python function with one variable $\mathtt{t}$. It contains the reserved word _return_. This statement tells the function to return the object after exectution of the function. This means that after execution of the function the function will return the value calculated in the variable $\mathtt{y}$.

One thing to notice here is that the exponential function is not part of the preloaded libraries in Python. That is one reason why we needed to import the numpy package. notice that the exponential function $\mathtt{np.exp}$ is one of the functions that were loaded with _numpy_.

We show next how to use this function.

```python
f(0)
```

This output can also be saved in a new variable

```python
out = f(3)

out
```

To plot the graph of a function what is done is to plot points in the xy-plane. One generates a list of values for $t$ where one wants to evaluate the function and ten plots points of the form $(t_i, f(t_i))$ on the plane. Typically, to give an appearance of smoothnes, these points are joined with straight lines.

The following command generates a vector (actually a numpy array) with the values of the variable $t$ where the function is going to be defined. The command _linspace_ generates a vector with equally spaced values with the determined starting and ending values. In particular the following expression returns a vector with 101 elements starting with the value $-3$ and ending in $3$.

```python
t = np.linspace(-3, 3, 101)

print(len(t)) # number of elements in the array
```

The numpy array is a type of object in Python that mimics the functionality of the Matlab array. You can think of the numpy array as a vector or matrix. It is essentially very similar to a list, but it allows to perform mathematical operations between different arrays like matrix multiplication.

```python
# The following command selects the first ten elements in the aray t

sample = t[0:10]

sample
```

Now that we have the values where we want to evaluate the function we can evaluate it. In this case it will as simple as the following statement

```python
y = f(t)

y[0:10]
```

This works because the numpy array, $\mathtt{t}$, admits mathematical operations and the expression in the definition of the function $\mathtt{f}$ contains valid operations among numpy arrays. This would not work with Python lists as the following code cells show.

## A simple plot

Now that we have defined a set of data we are ready to produce a plot.


```python
plt.plot(t,y)

plt.show()
```

The simple line above produced a plot. There are several commands that can be used to include more information in the figure. These are introduced below. Most of them are self explanatory. The way of introducing them is very similar to the Matlab syntax, it is possible that you are already familiar with it.

Notice that the figure below represents the same function. It appears to be different because the command $\mathtt{plt.axis}$ changed the limiting values of the axes displayed.

```python
plt.plot(t,y)


plt.title("My first plot in Jupyter")
plt.axis([0, 3, -0.1, 0.8]) # [x_min, x_max, y_min, y_max]
plt.xlabel("$t$")
plt.ylabel("$y$")
plt.legend(["$(t^2 + 1/2)\exp(-t^2)$"])
plt.grid()


plt.show()
```

## Understanding the process of creating a figure

With the set of staments above we created a _figure_. A _figure_ is the container of the plot. Each _figure_ may contain one or more _axes_, wich is where the actual plot of the function is drawn. To understand this better let us define a new function that we want to plot. Then we will try to plot the graphs of both functions in the same figure. There are several ways of doing this. 

```python
def f_2(t):

    y = (t**2+0.5)*np.exp(-t**2*1.5)
    
    return y

y_2 = f_2(t)
```

The first thing that one might try when willing to plot several functions is to use the above codes in a row. This will be done below. Compare the output of the following two cells. Notice that the only difference between them is that the command $\mathtt{plt.show}()$ after the first figure is commented out.

```python
plt.plot(t,y)

plt.grid()
plt.title("Plot of $(t^2 + 1/2)\exp(-t^2)$")
plt.xlabel("$t$")
plt.ylabel("$y$")
plt.legend(["$(t^2 + 1/2)\exp(-t^2)$"])

plt.show()

plt.plot(t,y_2)

plt.grid()
plt.title("Plot of $(t^2 + 0.5)\,\exp(-t^2*1.5)$")
plt.xlabel("$t$")
plt.ylabel("$y$")
plt.legend(["$(t^2 + 0.5)\,\exp(-t^2*1.5)$"])



plt.show()
```

```python
plt.plot(t,y)

plt.title("Plot of $(t^2 + 1/2)\exp(-t^2)$")
plt.xlabel("$t$")
plt.ylabel("$y$")
plt.legend(["$(t^2 + 1/2)\exp(-t^2)$"])
plt.grid()



# We have commented out the following command.
# This is the only difference with respect the previous cell. 
# plt.show() 

plt.plot(t,y_2)


plt.title("Plot of $(t^2 + 0.5)\,\exp(-t^2*1.5)$")
plt.xlabel("$t$")
plt.ylabel("$y$")
plt.legend(["$(t^2 + 0.5)\,\exp(-t^2*1.5)$"])
plt.grid()


plt.show()
```

To understand why we get this different behaviour we need to learn how the process of creating a figure is. 

The first time a command related with a figure is executed a new figure is created. All the subsequent commands are related to that figure. This is called a _block_ or _figure block_. The command $\mathtt{plt.show}()$ displays the figure and also closes it, so that any subsequent command related with a figure will create a new figure and include the output of that command in that new figure. That is why we got two figures in the first case and just one in the second.

In addition, notice that in the latter case the title and legend commands of the 'second' plot override those of the first one. Obviously one can have only one title on a figure, however the legend should represent all the elements plotted. More important, there is no grid displayed in this figure. A complete satisfactory example can be found below.



```python
plt.plot(t,y)

plt.plot(t,y_2,'r.')

plt.grid()
plt.title("Plot of $f(t)$ and $f_2(t)$")
plt.xlabel("$t$")
plt.ylabel("$y$")
plt.legend(["$(t^2 + 1/2)\exp(-t^2)$","$(t^2 + 0.5)\,\exp(-t^2*1.5)$"])



plt.show()
```

One important thing to know is that the command $\mathtt{plt.show}()$ is executed automatically at the end of each cell. This will close any open figures and display them. Check what happens in the next cell. A new figure is created. It has the default ticks for the axes and only contains a title. It is displayed automatically. ## Subplots and Axes. Beyond the Figure Blocks.

As stated before figures contain axes. In all the figures above there is only one set of axes. Some of the elements that we displayed in the figure are relative to the axes and some are relative to the figure itself. One needs to be aware of that if one wants to have a fine tuning of the figure.

The examples below are based on some of the examples found at the Gallery of the Matplotlib web page [Matplotlib Gallery](https://matplotlib.org/stable/gallery/index.html). Whenever you want to draw a figure it is wise to have a look here to have an idea of what type of figure is the most appropriate for your needs.

Below is an example with some randomly generated data that we will use to present some further constructions. In this example instead of _plot_ we are using _scatter_ which produces points in the picture. The colours of the points might be changed individually as well as their sizes. Try to modify the different parameters to see what changes you obtain. 

```python
N = 50
xs = np.random.rand(N)
ys = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2 
plt.scatter(xs, ys, s=area, c = colors, alpha=0.5)

plt.show()
```

Although the _figure block_ way of producing figures is useful, there is another way of modifying the figures. This second one is useful when one is working with many figures at a time and needs to change their parametes at once or needs to access a figure after another one is in use. This will also important when plotting several axes in just one figure. For instance, supose that we want to change the _xlabel_ property of several axes. We can do as before, but this implies that we have to type the $\mathtt{plt.xlabel()}$ statement once for each figure.

Instead of doing this we are going to safe a variable with the "axes handle" for each figure. The command used to do this is $\mathtt{plt.gca()}$ which stands for "Get Current Axes". There is a similar command to get the "handle" for a figure: $\mathtt{plt.gcf()}$.

```python
N = 50
xs = np.random.rand(N)*6-3
ys = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(xs, ys, s=area, c = colors, alpha=0.5)
ax1 = plt.gca() 

plt.figure()

plt.scatter(xs, -(ys-0.5)*2, s=area/2, c = colors, alpha=0.7)
ax2 = plt.gca()

for ax in [ax1, ax2]:
    ax.set_xlabel("$x$")
    ax.set(title="An Oak")

plt.show()
```

We have used the method $\mathtt{set}()$ to set the properties of the axes in a loop. The statement ax.set(title="An Oak") is equivalent to ax.set_title("An Oak"). Instead of the two staments in the for-loop one could have used:

> $\mathtt{ax.set(title="An\;Oak", xlabel="\$x\$")}$

Finally, as stated before, we can have several axes in one figure. This is done with the _subplots_ method. This method creates a figure with a certain number of axes associated and returns the handles of the figure itself and the axes so that they can be accessed.

When called with only one argument the axes are aligned vertically in the figure.

```python
fig, axs = plt.subplots(3)
fig.suptitle('Vertically stacked subplots')
axs[0].plot(t, y, "r+")
axs[1].plot(t, y_2,"g--")
axs[2].scatter(xs,ys,c = colors, alpha=0.7)


plt.show()
```

$\mathtt{fig}$ and $\mathtt{axs}$ are the variables that contain the "handles" of the figure and all the axes. The axes handle, $\mathtt{axs}$, is actually a numpy array whose elements are the different axes of the figure. To plot somethig in these axes we just have to proceed like in the previous figure examples with the scatter plots. That is, we use commands like

> axys_name.plot(t, y, "r+")

In this case axys_name must be replaced with the name of the variable that contains the corresponding handle.

Next we are going to stack the different axes in a matrix-like shape. The arguments of the _subplots_ command determine the shape by $\mathtt{plt.subplots(rows, columns)}$. 

In the next examples we will have two rows and two columns. In this case the numpy array containing the axes handle is a numpy array with two indices (row, and column).

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(t, y)
axs[0, 1].scatter(xs,ys, c = colors, alpha=0.7)
axs[1, 0].plot(t, -y, 'tab:green')
axs[1, 1].plot(t, -y_2, 'tab:red')

for i in range(2):
    for j in range(2):
        axs[i,j].set(xlabel='x-label', ylabel='y-label', title=f"Axis [{i},{j}]")
        axs[i,j].label_outer() # Displays labels only at the bottom and left axes.

        if j%2==0: # The Boolean expression is true for even numbers
            axs[i,j].grid()

plt.show()
```

There is a convenient way of iterating over all the axes without the need of a double for-loop. This is done with $\mathtt{axs.flat}$. This takes the two dimensional array ans transforms it in an iterable object that can be iterated in a single for-loop. 

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(t, y)
axs[0, 1].plot(t, y_2, 'tab:orange')
axs[1, 0].plot(t, -y, 'tab:green')
axs[1, 1].plot(t, -y_2, 'tab:red')

i = 0 # Initialising a counter to define the titles of the axes.

for ax in axs.flat:
    
    i += 1
    ax.set(xlabel='x-label', ylabel='y-label', title = f"Axis {i}")

# Displays labels only at the bottom and left axes.
for ax in axs.flat:
    ax.label_outer()
    
plt.show()
```
