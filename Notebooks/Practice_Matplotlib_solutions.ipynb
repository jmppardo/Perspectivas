##Â Exercise 2

Define two python functions that evaluate the following expressions:

1. $t\mapsto e^{-(t-1)^2/3}$
2. $x \mapsto x^2e^{-x}$



```python
def f_1(t):
    
    y = np.exp(-(t-1)**2/3)
    
    return y 

def f_2(x):
        
    y = x**2*np.exp(-x)
    
    return y
    
def f_3(x):
    
    y = np.cos(100*x)
    
    return y
```

## Exercise 3

Plot each of the functions defined in the previous exercise in separated figures. Use different styles for the lines of each figure. Give each figure a title, a legend and label the axes appropriately.

```python
t = np.linspace(-5,5,101)
y_1 = f_1(t)
x = np.linspace(0,5,1001)
y_2 = f_2(x)
y_3 = f_3(x)

plt.plot(t,y_1)
plt.title("First Figure")
plt.xlabel("t")
plt.ylabel("y")

plt.figure(2)

plt.plot(x,y_2, "k--")
plt.title("Second Figure")
plt.xlabel("$t_2$")
plt.ylabel("$\\theta$")
plt.legend("Figure")


plt.figure(3)

plt.plot(x,y_3, "r.")
```

## Exercise 8

Produce one figure with 6 axes aligned as a 3x2 matrix. The figure must have one title. Each set of axes in the figure must have a legend and a title. The labels must apear only at the bottom and the left margin of the figure. Use loops to minimise the statements that you need to complete this figure.

```python
fig, axs = plt.subplots(3,2)

fig.suptitle("Title of the figure")
fig.tight_layout()

y_1 = f_1(t)
y_2 = f_2(t)

y = [y_1, y_2, -y_1, -y_2, y_1 +1, y_2 -1]

i = 0

for ax in axs.flat: # This will iterate through the six elements of the axs array
    
    ax.plot(t, y[i])
    ax.set(xlabel = "$t$", title = f"Function $f_{i}$", ylabel = "y")
    ax.legend([f"$f_{i}$"])
    ax.label_outer()
    
    i += 1

plt.show()
```
