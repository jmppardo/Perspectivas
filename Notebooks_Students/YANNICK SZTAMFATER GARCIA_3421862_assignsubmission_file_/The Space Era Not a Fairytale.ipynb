### 1. Rockets, Satellites and Astronauts
**900s**: Gunpowdered arrows are thrown in China. You could call this the humble beginnings of rocketry. Or at least     projectiles. <br />
...<br />
**1700s**: Rockets see their use in the modern military, achieving altitudes of around 1.1km.<br />
**1903**: Russian scientist Konstantin Tsiolkovsky publishes key paper discussing rockets which use their own liquid fuel, oxidiser, possible orbits, stages…) shortly before writing his famous Tsiolkovsky equation: <br />
\begin{align}
\ ΔV = g_{0}·I_{SP}·ln(\frac{M_{INITIAL}}{M_{FINAL}}) 
\end{align}
 
**1926**: Robert Goddard (US): builds first liquid fuel rocket. Starts experimenting with steerable nozzles. People weren’t very fond of him initially, they thought he was a mad man: <br />
\begin{align}
\text{"Rocket science is no suitable pursuit for a physics professor!"}
\end{align} 
<br />
**1938**: Critical year. The V2 rocket is invented, under Wernher Von Braun´s supervision. Range 320km, altitude 90km. Altitude control, gyroscopes, military payload: a weapon. 4300 launched until 1945. <br />
**1945**: The end of the war, Russia takes the remaining rockets, the US takes the remaining brains: including Wehrner Von Braun. <br />
**1949**: Von Braun, once a top nazi rocket scientist, now is the Director of Marshall Space Flight Centre in Alabama, US. Tensions build in the following decade, giving rise to the Cold War, where the US and the USSR ultimately duel over world supremacy. <br />
**1957**: The start of the Space Age. The USSR successfully launches satellite Sputnik 1 into orbit, the first one of its kind.  And Sputnik 2, with Laika the dog, on it. <br />
**1958**: US launch Explorer 1, with a Geiger counter for cosmic ray flux. Space SCIENCE has begun.<br />
**1961**: USSR beat the US to another milestone: the first manned mission carrying Yuri Gagarin -Vostok 1, is launched into space. <br />
**1962**: US launch their first manned orbital mission, with John Glenn, onboard the Atlas. <br />
**1969**: (For completeness) The US carry out the largest space exploration endeavour yet, the Apollo 11 moon landing, where three astronauts set foot on the moon.

### 2. Space Globalisation
Sometime in between the last two events in the timeline, the international community picked up on the space trend, and space agencies began to be formed, funding and promoting numerous kinds of research and exploration missions into space. 
And of course the commercialization of space could not wait. Technology developments in distinct areas lead to an exponential rise in satellite deployment. Nowadays there are an estimated 6000 to 7000 satellites in orbit. Next you will guess how many of those are actually operational.

```python
print("What´s the number of Operational Satellites out of those in orbit?")
N_Op_Sats = wg.IntSlider(
    min=0,
    max=8000,
    step=1,
    description='Op. Sats:',
    continuous_update=True,
    orientation='horizontal',
    readout=True,
)
button = wg.Button(description="Go")
output = wg.Output()

def showOutput(btn):
    output.clear_output()
    return_value = N_Op_Sats.value
    with output:
        print(return_value)

button.on_click(showOutput)
ui = wg.HBox([N_Op_Sats, button, output])
ui

def opsats(x = N_Op_Sats):
    if x == 2700:
        print("Correct",x)
    elif x < 2000 and x > 1500:
        print("Incorrect, slightly more")
    elif x > 3000 and x < 4000:
        print("Incorrect, slightly less")
    elif x < 1500:
        print("Incorrect, significantly more")
    elif x > 4000:
        print("Incorrect, significantly less")
    else:
        print("Correct, around",x)
    return
Int = wg.interact(opsats);
```

### 3. Predictions of satellite numbers forwards in time
As mentioned, today, thousands of satellites remain in orbit around the Earth, spread in pre-defined orbits to suit their purpose. This number will not but keep increasing, probably exponentially, up to a point where certain regions of space could become inoperable due to the high likelihood of collision. This is called the Kessler Syndrome.

It is not only operational satellites that are a problem, but inert or dead ones too, and even fragments of satellites. Often times when satellite parts do not decay into the Earth and re-enter the atmosphere (to be burned up), they remain in orbit, being an absolute waste of …space. Not only that, collisions between satellites are possible, and have the potential to create thousands of new, tiny fragments, which can now act as bullets to any other satellite passer-by. Right now there are around 25,000 OBJECTS in orbit, larger than 10cm. Over 600,000 larger than 1cm.
